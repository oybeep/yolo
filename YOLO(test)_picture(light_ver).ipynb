```python
from ultralytics import YOLO
from matplotlib import pyplot as plt
import cv2
```

```python
model = YOLO("yolov8s.pt")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 입력 영상
input_video = "/content/drive/MyDrive/third_floor_corridor.mp4"
cap = cv2.VideoCapture(input_video)

# 출력 영상 세팅
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
output_path = "/content/drive/MyDrive/output_detected_person.mp4"

# 원본 FPS 유지
fps = cap.get(cv2.CAP_PROP_FPS)

# 원하는 출력 크기 (프레임 축소), 예: 640x360
resize_width = 640
resize_height = 360

out = cv2.VideoWriter(output_path, fourcc, fps, (resize_width, resize_height))

# 프레임 스킵 값 (예: 3 → 3프레임 중 1프레임만 처리)
FRAME_SKIP = 3
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 프레임 스킵
    if frame_count % FRAME_SKIP != 0:
        frame_count += 1
        continue

    # 프레임 크기 줄이기
    frame_resized = cv2.resize(frame, (resize_width, resize_height))

    # YOLO 탐지
    results = model.predict(
        source=frame_resized,
        conf=0.5,
        imgsz=640,       # YOLO 내부 입력 크기
        verbose=False     # 출력 줄이기
    )

    # 박스 그리기
    annotated_frame = results[0].plot()

    # 저장
    out.write(annotated_frame)

    frame_count += 1

cap.release()
out.release()

print("완료! 결과 영상 저장됨:", output_path)
```
